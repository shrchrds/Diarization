```python
speaker_array = []
start_array = []
end_array = []
```

```python
# print the result
for turn, _, speaker in diarization.itertracks(yield_label=True):
    speaker_array.append(speaker)
    start_array.append(round(turn.start,2)*1000) # pydub calculates in millisec
    end_array.append(round(turn.end,2)*1000) # pydub calculates in millisec
    print(f"start={turn.start:.1f}s stop={turn.end:.1f}s speaker_{speaker}")
```

```python
from pydub import AudioSegment
```

```python
# split audio using timeline
for filename in os.listdir("/content/data"):
    print("Printing filename :", filename)
    save_file_name = filename[:-4]
    myaudio = AudioSegment.from_file("data"+"/"+filename, "wav")

    for i in range(len(start_array)):
        chunk_name = save_file_name + f"_{i}.wav"
        parts = chunk_name.split('_')
        a = parts[-1].split('.')
        b = a[0]
        if(len(b) == 1):
            b = "0"+b
        chunk_name = parts[0] + "_" + parts[1] + "_" + b + "_" + speaker_array[i]+ ".wav"
        print("-- Printing chunk_name : ",chunk_name)

        chunk_data = myaudio[start_array[i]: end_array[i]]
        chunk_data.export("outputdir"+"/" + chunk_name, format="wav")
```

```python
import os
from faster_whisper import WhisperModel
```

```python
model_size = "small"
model = WhisperModel(model_size, device="cpu", compute_type="int8")
```

```python
def transcribe_audio(model, audio_path):
    segments, info = model.transcribe(audio_path, beam_size=5)
    transcribed_text = ""
    for segment in segments:
        print("[%.2fs -> %.2fs] %s" % (segment.start, segment.end, segment.text))
        transcribed_text += segment.text

    return transcribed_text
```

```python
transcription = transcribe_audio(model, "/content/data/test.wav")
```
